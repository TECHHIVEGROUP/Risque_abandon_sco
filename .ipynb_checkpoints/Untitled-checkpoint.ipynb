### 1. Compréhension du problème (Business Understanding)
L’abandon scolaire est un enjeu majeur pour les établissements d’enseignement. Il entraîne non seulement des conséquences sociales pour l’étudiant, mais aussi des pertes de ressources pour l’institution. L’objectif de ce projet est de construire un modèle de Machine Learning capable de prédire, à partir de données académiques et comportementales, si un étudiant présente un risque élevé d’abandon. Il s’agit d’un problème de classification binaire où la variable cible dropout_risk prend la valeur 1 (risque d’abandon) ou 0 (pas de risque). La mise en place d’un tel outil nous permettrait de déclencher des actions préventives personnalisées.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, confusion_matrix, classification_report)
```

### 2. Chargement des données
Le jeu de données utilisé contient plusieurs centaines d’étudiants décrits par des variables variées : âge, sexe, moyenne générale, taux d’absentéisme, accès à Internet, temps d’étude hebdomadaire et participation à des activités extrascolaires. La variable cible a été construite artificiellement selon une règle métier : un étudiant est considéré à risque si au moins deux des trois conditions suivantes sont réunies (moyenne < 10, absentéisme > 30%, temps d’étude < 1h). Avant toute modélisation, une première exploration s’impose pour comprendre la structure des données, détecter d’éventuelles anomalies et visualiser la distribution des variables.

```python
df = pd.read_csv("./data/student_dropout_dataset.csv")
df.head()
```

### 3. Encodage et normalisation
Les données brutes ne sont pas toujours exploitables directement par les algorithmes de Machine Learning. Certaines variables sont catégorielles (sexe, accès Internet, activités extrascolaires) et doivent être converties en valeurs numériques via un encodage adapté (One-Hot Encoding). Par ailleurs, les variables numériques présentent des échelles différentes (âge, notes, taux d’absence) ; une normalisation (Z-score) est donc appliquée pour éviter que certaines variables ne dominent artificiellement le modèle. Enfin, nous vérifions l’absence de valeurs manquantes dans ce dataset.

```python
# Séparation features / target
X = df.drop("dropout_risk", axis=1)
y = df["dropout_risk"]
```

```python
# Colonnes numériques et catégorielles
num_cols = ["age", "average_grade", "absenteeism_rate", "study_time_hours"]
cat_cols = ["gender", "internet_access", "extra_activities"]
```

```python
# Preprocessing pipeline
preprocessor = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ("cat", OneHotEncoder(drop="first"), cat_cols)
])
```

### 4. ANALYSE EXPLORATOIRE (EDA)
L’analyse exploratoire des données (EDA) permet de visualiser les relations entre les variables et d’identifier celles qui influencent le plus la variable cible. Nous utilisons pour cela des histogrammes pour observer la distribution de chaque variable numérique, une matrice de corrélation pour détecter d’éventuelles dépendances linéaires, ainsi que des boxplots comparant la répartition de chaque variable selon que l’étudiant est à risque ou non. Cette étape guide ensuite le choix des features à privilégier.

```python
# Distribution de la variable cible
print("\nDistribution de dropout_risk :")
print(df["dropout_risk"].value_counts())
print(f"Pourcentage d'abandon : {df['dropout_risk'].mean()*100:.1f}%")
```

```python
# Histogrammes
df.hist(figsize=(12, 8))
plt.suptitle("Histogrammes des variables numériques", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Matrice de corrélation
plt.figure(figsize=(8, 6))
sns.heatmap(df[num_cols + ["dropout_risk"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
```

```python
# Boxplots pour identifier les variables influentes
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x="dropout_risk", y=col, ax=axes[i//2, i%2])
plt.tight_layout()
plt.show()
```

### 5. FEATURE ENGINEERING
Pour améliorer les performances prédictives du modèle, nous créons de nouvelles variables à partir de celles existantes. Ainsi, nous définissons un ratio présence/absence (complément du taux d’absentéisme) et un score global qui combine la moyenne générale, la présence et le temps d’étude. Ces nouvelles caractéristiques peuvent capturer des comportements plus riches que les variables brutes. Enfin, une sélection des variables les plus importantes sera effectuée à l’issue de la modélisation.

```python
df["presence_ratio"] = 1 - df["absenteeism_rate"]
df["effort_score"] = df["study_time_hours"] * df["average_grade"] / 20
df["global_score"] = (df["average_grade"] * 0.5 + 
                      df["presence_ratio"] * 20 * 0.3 + 
                      df["study_time_hours"] * 2 * 0.2)
```

```python
# Mise à jour des features
X_enhanced = df.drop("dropout_risk", axis=1)
num_cols_enhanced = ["age", "average_grade", "absenteeism_rate", "study_time_hours", 
                     "presence_ratio", "effort_score", "global_score"]
```

```python
X_enhanced
```

```python
preprocessor_enhanced = ColumnTransformer([
    ("num", StandardScaler(), num_cols_enhanced),
    ("cat", OneHotEncoder(drop="first"), cat_cols)
])
```

```python
preprocessor_enhanced
```

### 6. SPLIT TRAIN/TEST
Pour évaluer objectivement la capacité de généralisation des modèles, nous divisons le jeu de données en deux ensembles : 80 % pour l’entraînement et 20 % pour le test. Cette séparation est stratifiée afin de conserver la même proportion de la classe cible dans les deux échantillons. En complément, une validation croisée à 5 plis (5-fold cross-validation) est appliquée sur l’ensemble d’entraînement pour estimer la stabilité des performances et éviter le sur-apprentissage.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_enhanced, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
print(f"\nTaille train : {X_train.shape}")
print(f"Taille test : {X_test.shape}")
```

```python
X_train
```

### 7. MODÉLISATION
Trois algorithmes de classification supervisée sont testés dans ce projet : la régression logistique (modèle linéaire simple et interprétable), la forêt aléatoire (Random Forest, modèle ensembliste non linéaire robuste) et les machines à vecteurs de support (SVM, efficaces en haute dimension). Chaque modèle est évalué sur sa capacité à distinguer les étudiants à risque de ceux qui ne le sont pas. L’objectif est de comparer leurs performances afin de sélectionner le plus pertinent pour notre problème.

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42)
}
```

```python
results = {}
for name, model in models.items():
    pipeline = Pipeline([
        ("prep", preprocessor_enhanced),
        ("clf", model)
    ])
    
    # Validation croisée
    cv_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring="f1")
    
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    
    results[name] = {
        "CV F1 moyen": cv_scores.mean(),
        "Test Accuracy": accuracy_score(y_test, y_pred),
        "Test Precision": precision_score(y_test, y_pred),
        "Test Recall": recall_score(y_test, y_pred),
        "Test F1": f1_score(y_test, y_pred)
    }
    
    print(f"\n{'='*50}")
    print(f"Modèle : {name}")
    print(f"{'='*50}")
    print(f"Validation croisée F1 (moy ± std) : {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
    print(f"\nClassification Report sur test :")
    print(classification_report(y_test, y_pred))
    print("Matrice de confusion :")
    print(confusion_matrix(y_test, y_pred))
```

```python
print("\n" + "="*60)
print("COMPARAISON DES MODÈLES")
print("="*60)
comparison_df = pd.DataFrame(results).T
print(comparison_df.round(3))
```

```python
print("\n" + "="*50)
print("OPTIMISATION - Random Forest")
print("="*50)

param_grid = {
    "clf__n_estimators": [50, 100, 200],
    "clf__max_depth": [5, 10, None],
    "clf__min_samples_split": [2, 5, 10],
    "clf__min_samples_leaf": [1, 2, 4]
}

rf_base = RandomForestClassifier(random_state=42)
pipeline_rf = Pipeline([("prep", preprocessor_enhanced), ("clf", rf_base)])

grid_search = GridSearchCV(
    pipeline_rf, param_grid, cv=5, scoring="f1", n_jobs=-1, verbose=1
)
grid_search.fit(X_train, y_train)

print(f"\nMeilleurs paramètres : {grid_search.best_params_}")
print(f"Meilleur score F1 (validation) : {grid_search.best_score_:.3f}")
```

```python
# Évaluation du modèle optimisé
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)
print(f"F1 score sur test : {f1_score(y_test, y_pred_best):.3f}")
print(f"Accuracy sur test : {accuracy_score(y_test, y_pred_best):.3f}")
```

```python
# SÉLECTION DES VARIABLES IMPORTANTES 
best_rf = best_model.named_steps["clf"]
feature_names = (num_cols_enhanced + 
                 list(best_model.named_steps["prep"].named_transformers_["cat"].get_feature_names_out(cat_cols)))

importance_df = pd.DataFrame({
    "feature": feature_names,
    "importance": best_rf.feature_importances_
}).sort_values("importance", ascending=False)

print("\nTop 5 des variables les plus importantes :")
print(importance_df.head(5))
```

```python
# Visualisation des importances
plt.figure(figsize=(10, 6))
plt.barh(importance_df["feature"][:8], importance_df["importance"][:8])
plt.xlabel("Importance")
plt.title("Importance des variables (Random Forest optimisé)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# SAUVEGARDE DU MODÈLE 
best_params_clean = {}
for key, value in grid_search.best_params_.items():
    clean_key = key.replace("clf__", "")
    best_params_clean[clean_key] = value

print("Paramètres nettoyés :", best_params_clean)

# Créer le modèle final avec les bons paramètres
final_model = Pipeline([
    ("prep", preprocessor_enhanced),
    ("clf", RandomForestClassifier(**best_params_clean, random_state=42))
])

# Entraîner le modèle final
final_model.fit(X_train, y_train)

# Sauvegarder
# Creation du dossier
os.makedirs("./models", exist_ok=True)

# Sauvegarde du modele final
joblib.dump(final_model, "./models/dropout_model.pkl")
print(f"Modele sauvegarde dans ./models/dropout_model.pkl")
print(f"Fichier existe: {os.path.exists('./models/dropout_model.pkl')}")
```

```python
# PRÉDICTION SUR UN NOUVEL ÉTUDIANT 
print("\n" + "="*50)
print("TEST SUR UN NOUVEL ÉTUDIANT")
print("="*50)

new_student = pd.DataFrame({
    "age": [19],
    "gender": ["Male"],
    "average_grade": [9.5],
    "absenteeism_rate": [0.35],
    "internet_access": ["No"],
    "study_time_hours": [0.8],
    "extra_activities": ["No"],
    "presence_ratio": [0.65],
    "effort_score": [9.5 * 0.8 / 20],
    "global_score": [0]
})
new_student["global_score"] = (new_student["average_grade"] * 0.5 + 
                               new_student["presence_ratio"] * 20 * 0.3 + 
                               new_student["study_time_hours"] * 2 * 0.2)

prediction = final_model.predict(new_student[num_cols_enhanced + cat_cols])
proba = final_model.predict_proba(new_student[num_cols_enhanced + cat_cols])

print(f"Risque d'abandon : {'OUI' if prediction[0] == 1 else 'NON'}")
print(f"Probabilité d'abandon : {proba[0][1]:.2%}")
```
